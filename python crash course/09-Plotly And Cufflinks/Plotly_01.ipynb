DUE TO USE OF PLOTLY LIBRARY IT DIRECTLY CREATES STUNNING PLOTS JUST BY USING .plot() METHOD

```python
df.plot()
```

TO MAKE IT MORE STUNNING WE USE .iplot() METHOD

```python
df.iplot()
```

TO CREATE SCATTER PLOT WE USE iplot(kind='scatter'x=,y=)

```python
df.iplot(kind='scatter',x='A',y='B')
```

```python
df.iplot(kind='scatter',x='A',y='B', mode='markers')
```

TO CHANGE THE SIZE OF THE MARKERS WE USE size=20

```python
df.iplot(kind='scatter', x='A',y='B', mode='markers', size=20)
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.iplot(kind='bar')
```

WE ADD SOME AGGREGATE FUNCTION BEFORE USING THE BAR ON LARGE DATAS

```python
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

```python
df2.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,4000,300,200,100]})
df3.iplot(kind='surface')
```

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3.iplot(kind='surface')
```

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,4000,300,200,100]})
df3.iplot(kind='surface',colorscale='rdylbu')
```
