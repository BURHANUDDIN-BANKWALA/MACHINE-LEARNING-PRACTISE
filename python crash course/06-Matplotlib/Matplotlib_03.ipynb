COLORS

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
ax.plot(x,y,color='red')
fig
```

```python
ax.plot(x,y,color='green')
fig
```

```python
ax.plot(x,y,color='purple')
fig
```

```python
""" ax.plot(x,y,color='##FFF000')
fig """
```

LINE WIDTH

```python
ax.plot(x,y,linewidth=4)
fig
```

```python
ax.plot(x,y,alpha=0.5)
fig
```

LINE STYLE

```python
ax.plot(x,y,color='red',linestyle='--')
fig
```

```python
ax.plot(x,y,color='red',linestyle=':',linewidth=5)
fig
```

```python
ax.plot(x,y,color='red',linestyle='-',linewidth=6)
fig
```

MARKERS


```python
ax.plot(x,y,marker='o')
fig
```

```python
ax.plot(x,y,marker='+',markersize=4,markerfacecolor='yellow', markeredgewidth=4, markeredgecolor='g')
fig
```

SET X AND Y LIMIT

```python
ax.set_xlim([0,1])
ax.set_ylim([0,2])
fig
```
