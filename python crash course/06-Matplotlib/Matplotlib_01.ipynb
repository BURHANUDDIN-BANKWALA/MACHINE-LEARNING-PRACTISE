FUNCTIONAL METHOD TO PLOT A GRAPH

FROM plt WE WILL CALL plt.plot() METHOD

```python
plt.plot(x,y)
```

TO VISUALIZE WE WILL TYPE plt.show()

REMEMBER WE CAN USE plt.show ONLY ONCE FOR ONE PLOT IF ANOTHER PLOT IS CREATED THEN plt.show WILL SHOW IT AND ALL THE UPDATE WILL GO INTO IT

TO GIVE LABELS WE WILL USE plt.xlabel('x label') AND plt.ylabel('y label')

```python
plt.xlabel('X LABEL')
```

```python
plt.ylabel('Y LABEL')
```

FOR TITLE WE WILL USE plt.title('TITLE')

```python
plt.title('TITLE')
```

```python
plt.show()
```

MULTIPLOTTING


```python
plt.plot(x,y,'r')
```

```python
plt.show()
```

FOR SUBPLOTTING WE USE plt.subplot(no.of rows, no. of columns, plot no)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.show()
```

OBJECT ORIENTED METHOD

WE WILL FIRST CREATE A BLANK SHEET USING plt.figure()

```python
fig=plt.figure()
```

THEN WE WILL DEFINE AXES USING fig.add_axes([left,bottom,height,width])

```python
axes=fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
axes.plot(x,y)
```

```python
plt.show()
```

WHEN USING OBJECT ORIENTED METHOD WE USE .set_xlabel() TO SET THE LABEL

```python
axes.set_xlabel('X LABEL')
```

```python
axes.set_ylabel('Y LABEL')
```

```python
axes.set_title('TITLE')
```

```python
plt.show()
```

```python
fig=plt.figure()
```

THE AXES CAN ONLY LIE FROM 0 TO 1

```python
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.3,0.4,0.2,0.2])
```

```python
axes1.plot(x,y)
axes1.set_title('Larger plot')
```

```python
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
plt.show()
```
