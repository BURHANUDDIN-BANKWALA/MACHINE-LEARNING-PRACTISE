JOIN PLOT

TO USE JOIN PLOT WE USE sns.joinplot(x='column1', y='column2', data=database)


```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

WE CAN CUSTOMIZE THE JOINTPLOT BY sns.jointplot(x=, y=, data=, kind='hex') kind='reg' or kind='kde'

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

PAIR PLOT

PAIR PLOT GIVES AN OVERVIEW WITH DIFFERENT COLUMNS

```python
sns.pairplot(tips)
```

WE USE HUE TO HIGHLIGHT DIFFERENT TYPES

WE DO THIS BY sns.pairplot(data, hue='column name')

```python
sns.pairplot(tips, hue='sex')
```

TO CHANGE COLOR IN THE PLOT WE USE sns.pairplot(data, hue='column_name', palette='coolwarm')

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

RUG PLOT

SIMILIAR TO DISTPLOT RUG PLOT ALSO TAKES ONLY ONE VARIABLE
THE RUG PLOT SHOWS DENSITY 

```python
sns.rugplot(tips['total_bill'])
```

KDEPLOT
IT ALSO TAKES ONLY ONE VARIABLE AND SHOWS DISTRIBUTION OR DENSITY


```python
sns.kdeplot(tips['total_bill'])
```
