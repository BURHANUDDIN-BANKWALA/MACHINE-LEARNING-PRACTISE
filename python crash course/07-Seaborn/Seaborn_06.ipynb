STYLE AND COLOR

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

TO CUSTOMIZE THE BACKGROUND BEHIND THE PLOTS WE USE sns.set_style() METHOD 


```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

WE CAN ALSO REMOVE EXTRA LINES BY USING sns.despine() METHOD

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

TO CHANFE THE PLOT SIZE MANUALLY WE USE MATPLOTIBS plt.figure(figsize= ()) REMEMBER TO PASS IN TUPLE

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(13,3))
sns.countplot(x='sex',data=tips)
```

TO CHANGE THE GENERAL ASPECT RATIO BY USING sns.set_context() TO poster, notebook, talk AND paper.

```python
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('talk')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('paper')
sns.countplot(x='sex',data=tips)
```
