CATAGORICAL PLOTS

```python
import seaborn as sns
import numpy as np
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

BAR PLOTS

BARPLOTS TAKE X AXIS AS CATEGORICAL LIKE SEX OR DAY AND Y AXIS AS NUMERICAL

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

WE CAN USE AGGREGATE FUNCTIONS IN BAR PLOTS BY DEFAULT AVERAGE IS SET. 
sns.barplot(x=, y=, data=, estimator=* aggregate function*)

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

COUNT PLOTS

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='smoker', data=tips)
```

VIOLIN PLOT

```python
sns.violinplot(x='sex', y='tip', data=tips)
```

WE CAN USE hue='column_name' TO SEE THE DIFFERENCE

```python
sns.violinplot(x='day', y='tip', hue='sex', data=tips)
```

WE CAN MERGE THE HUE BY ENABLING THE split=True


```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

STRIP PLOT

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

TO VISUALIZE EACH POINT MORE EASILY WE ENABLE JITTER TO SEE THE DATA MORE CLEARLY 

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

BOX PLOTS OR WHISKER PLOTS

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

SWARM PLOT

DO NOT USE IT FOR VERY LARGE DATASETS

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```
