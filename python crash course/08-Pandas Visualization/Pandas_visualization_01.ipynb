IN ORDER TO CREATE PLOTS FAST WE USE PANDAS VISUALIZATION INSTEAD OF SEABORN.
WE USE SEABORN FOR SOFISTICATED DISPLAY OF OUR DATA.


```python
import numpy as np
import pandas as pd
df1=pd.read_csv('df1')
df1.head()
```

```python
df2=pd.read_csv('df1',index_col=0)
df2.head()
```

```python
df1['A'].count()
```

INSTEAD OF USING MATPLOTLIB OR SEABORN WE CAN DIRECTLY CREATE PLOTS OUT OF PANDAS

```python
df1['A'].hist()
```

```python
df1['A'].hist(bins=30)
```

BUT WE CAN ENHANCE THEM JUST BY IMPORTING THE SEABORN LIBRARY IT DOES THE MAGIC FOR US

```python
import seaborn as sns 
df1['A'].hist(bins=30)
```

ANOTHER WAY OF MAKING HIST PLOT IS BY USING .plot(kind='kde') METHOD

```python
df1['A'].plot(kind='hist', bins=50)
```

```python
df1['A'].plot(kind='kde')
```

AREA PLOT


```python
arr=np.random.randint(1,50000,2000).reshape(1000,2)
arr=pd.DataFrame(arr, columns=['a','b'])
```

```python
arr.plot.area(alpha=0.4)
```

```python
arr.plot.area(alpha=0.4, stacked=True)
```

SCATTER PLOT


```python
df1.plot.scatter(x='A',y='B')
```

WE CAN ALSO ADD A THIRD PARAMETER AS COLOUR c='column_name'

```python
df1.plot.scatter(x='A',y='B',c='C')
```

SIMILARILY WE CAN ADD SIZE AS s=df['column_name']*30

```python
df1.plot.scatter(x='A',y='B', s=df1['C']*30)
```

WE CAN CHANGE COLOR BY USING cmap='coolwarm'

```python
df1.plot.scatter(x='A',y='B',c='C', cmap='coolwarm')
```
